```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
plt.style.use('seaborn-whitegrid')
%matplotlib inline


df = pd.read_csv('/Users/congliu/Downloads/qc.result.xls', sep='\t')
df.head()
```

```python
fig, ax = plt.subplots(2)
x = df['sample']
y = df['rawreads']
y2 = df['rawbases']

ax[0].plot(x, y, '--')
ax[0].set_xlabel('sample')
ax[1].plot(x, y2, '--')
ax[1].set(xlabel = 'sample',
          ylabel = 'rawbases',
          title = 'rawbases')
plt.show()
```

```python
sns.set(style = 'whitegrid')
ax = sns.lineplot(data=df, x='sample', y='rawreads')
# [method_name for method_name in dir(ax) if callable(getattr(ax, method_name))]
ax.lines[0].set_linestyle("--")

plt.show()
```

```python
f,a = plt.subplots()
a.plot(df['sample'], df['rawreads'], linestyle='--')
a.set(xlabel='sample',
      ylabel='reads',
      title='test')
plt.xticks(rotation='45')
plt.savefig("/Users/congliu/Downloads/aa.png", dpi=800)
plt.show()
plt.close()
```
